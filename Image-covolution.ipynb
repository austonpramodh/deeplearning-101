## Edge Detections with convolutions

```python
# Read image
url="https://cdn.pixabay.com/photo/2018/03/31/06/31/dog-3277416_960_720.jpg"
from PIL import Image
import requests
from io import BytesIO

# response = requests.get(url)
# img = Image.open(BytesIO(response.content))
img = Image.open("./profile_pic.jpg")
img_size = 300
img = img.resize((img_size, img_size))
img = ImageOps.grayscale(img)
print("Actual Image!!")
display(img)
imgArray = np.array(img)
```

```python
# Vertical Edge Detection
vertical_kernal = [
    [1, 0, -1],
    [1, 0, -1],
    [1, 0, -1]
]
# Convolute the image
conv_img = conv2D(vertical_kernal, imgArray)
pixels = np.array(conv_img)

image = Image.fromarray(pixels.astype('uint8'))

print("Vertical Edge Detected Image")
display(image)

# Horizontal Edge Detection
horizontal_kernal = [
    [1, 1, 1],
    [0, 0, 0],
    [-1, -1, -1]
]
```

```python
# Convolute the image
conv_img = conv2D(horizontal_kernal, imgArray)
pixels = np.array(conv_img)

image = Image.fromarray(pixels.astype('uint8'))
print("Horizontal Edge Detected Image")
display(image)
```

```python
# Horizontal Edge Detection
horizontal_kernal = [
    [1, 0, -1],
    [2, 0, -2],
    [1, 0, -1]
]
# Convolute the image
conv_img = conv2D(horizontal_kernal, imgArray)
pixels = np.array(conv_img)

image = Image.fromarray(pixels.astype('uint8'))
print("Sobel Edge Detected Image")
display(image)
```
